# **1. Data Pre-processing**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
'''%matplotlib inline'''
```

```python
file_id = '1eGPfa3tLQ6uilh9znshF9qaQGHlqY1Wd'
link = f'https://drive.google.com/uc?id={file_id}'
bank_data = pd.read_csv(link)
bank_data
```

```python
bank_data.isnull().sum()
```

```python
c_impute = SimpleImputer(missing_values= np.nan, strategy = 'most_frequent')
bank_data[['Mortgage','Personal Loan','Securities Account']] = c_impute.fit_transform(bank_data[['Mortgage','Personal Loan','Securities Account']])
```

```python
bank_data.isnull().sum()
```

```python
plt.figure(figsize=(15,8))
plt.title("Heatmap showing Correlation between all the features", fontsize=20)
sns.heatmap(bank_data.corr(),annot = True, cmap='mako')
```

```python
df1 = bank_data.drop(["ID","ZIP Code"], axis = 1)
df1
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled = scaler.fit(df1.drop('CreditCard',axis=1)).transform(df1.drop('CreditCard',axis=1))
df_scaled = pd.DataFrame(scaled, columns=df1.columns[:-1])
df_scaled.head()
```

```python
x = df_scaled
y = df1['CreditCard']

x.head()
```

```python
y.head()
```

# Use SVM (Support Vector Machines)

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
svc=SVC()
svc.fit(x_train, y_train)
y_pred=svc.predict(x_test)
print('Model accuracy : {0:0.3f}'. format(accuracy_score(y_test, y_pred)))
```

```python
rbf_svc=SVC(kernel='rbf').fit(x_train,y_train)
y_pred = rbf_svc.predict(x_test)
print('Model accuracy with rbf kernel : {0:0.3f}'. format(accuracy_score(y_test, y_pred)))
```

```python
predicted = svc.predict(x_test)
```

```python
print('Accuracy =', svc.score(x_test,y_test)*100)
```
